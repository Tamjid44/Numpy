```python
import pandas as pd
import numpy as np
index = 'r1 r2 r3 r4 r5 r6 r7 r8 r9 r10'.split()
columns = 'c1 c2 c3 c4 c5 c6 c7 c8 c9 c10'.split()
df = np.arange(100,0,-1).reshape(10,10)
df = pd.DataFrame(data=df,index=index,columns=columns)
print(df)
```

**df** is our first dataframe. <br>

Each column is actually a pandas series, sharing a common index (row labels). <br>

&#9758; Let's learn how to **Grab data** that we need, this is the most important thing we want to learn to move one!<br>

### Grab a single column c10

```python
import pandas as pd
import numpy as np
index = 'r1 r2 r3 r4 r5 r6 r7 r8 r9 r10'.split()
columns = 'c1 c2 c3 c4 c5 c6 c7 c8 c9 c10'.split()
df = np.arange(100,0,-1).reshape(10,10)
df = pd.DataFrame(data=df,index=index,columns=columns)
df = df['c10']
print(df)
```

### Grab multiple columns c4, c2, c9

```python
import pandas as pd
import numpy as np
index = 'r1 r2 r3 r4 r5 r6 r7 r8 r9 r10'.split()
columns = 'c1 c2 c3 c4 c5 c6 c7 c8 c9 c10'.split()
df = np.arange(100,0,-1).reshape(10,10)
df = pd.DataFrame(data=df,index=index,columns=columns)
df = df [['c4','c2','c9']]
print(df)
```

### Add a new column name 'new' and value would be c1 * c10

```python
import pandas as pd
import numpy as np
index = 'r1 r2 r3 r4 r5 r6 r7 r8 r9 r10'.split()
columns = 'c1 c2 c3 c4 c5 c6 c7 c8 c9 c10'.split()
df = np.arange(100,0,-1).reshape(10,10)
df = pd.DataFrame(data=df,index=index,columns=columns)
df['new']= df['c1']*df['c10']
print(df)
```

### Delete the column 'new' permanently  

```python
import pandas as pd
import numpy as np
index = 'r1 r2 r3 r4 r5 r6 r7 r8 r9 r10'.split()
columns = 'c1 c2 c3 c4 c5 c6 c7 c8 c9 c10'.split()
df = np.arange(100,0,-1).reshape(10,10)
df = pd.DataFrame(data=df,index=index,columns=columns)
df['new']= df['c1']*df['c10']
df = df.drop(columns=['new'])
print(df)
```

### Grabb more than one rows 'r2', 'r1','r3'

```python
import pandas as pd
import numpy as np
index = 'r1 r2 r3 r4 r5 r6 r7 r8 r9 r10'.split()
columns = 'c1 c2 c3 c4 c5 c6 c7 c8 c9 c10'.split()
df = np.arange(100,0,-1).reshape(10,10)
df = pd.DataFrame(data=df,index=index,columns=columns)
df = df.loc[['r2','r1','r3']]
print(df)
```

### Grab a sub-set of the dataframe where c1 is less than 50

```python
import pandas as pd
import numpy as np
index = 'r1 r2 r3 r4 r5 r6 r7 r8 r9 r10'.split()
columns = 'c1 c2 c3 c4 c5 c6 c7 c8 c9 c10'.split()
df = np.arange(100,0,-1).reshape(10,10)
df = pd.DataFrame(data=df,index=index,columns=columns)
df = df[df['c1']<50]
print(df)
```

### Grab a sub-set of the dataframe where c3 is equal to 38 and c8 is equal to 33

```python
import pandas as pd
import numpy as np
index = 'r1 r2 r3 r4 r5 r6 r7 r8 r9 r10'.split()
columns = 'c1 c2 c3 c4 c5 c6 c7 c8 c9 c10'.split()
df = np.arange(100,0,-1).reshape(10,10)
df = pd.DataFrame(data=df,index=index,columns=columns)
df = df[(df['c3']==38) & (df['c8']==33)]
print(df)
```
